**YData Profiling is an automated exploratory data analysis (EDA) library that generates an interactive HTML report summarizing dataset structure, variable distributions, missing values, correlations, outliers, and data quality issues. It helps quickly understand the data and make informed decisions for feature engineering and model building.**

```python
import pandas as pd
from ydata_profiling import ProfileReport

df = pd.read_csv("gurgaon_properties_cleaned_v2.csv").drop_duplicates()

profile = ProfileReport(
    df,
    title="Automated EDA Report using YData Profiling",
    explorative=True
)

profile.to_file("output_report.html")
```
